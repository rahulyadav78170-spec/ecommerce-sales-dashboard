```python
import pandas as pd
 
# ── Step 1: Load Data ──────────────────────────────────────────
df = pd.read_csv('Amazon Sale Report.csv', encoding='unicode_escape')
print('Shape:', df.shape)
print('Columns:', df.columns.tolist())
print('Missing Values:')
print(df.isnull().sum())
 
# Step 2: Clean Column Names
df.columns = df.columns.str.strip()
 
# Step 3: Drop Useless Columns 
df.drop(columns=['index', 'Unnamed: 22'], inplace=True, errors='ignore')

# Step 4: Fix Date Column 
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df.dropna(subset=['Date'], inplace=True)   # remove rows with no date
 
# Step 5: Fix Numeric Columns 
df['Qty'] = pd.to_numeric(df['Qty'], errors='coerce').fillna(0).astype(int)
df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce').fillna(0)
 
#  Step 6: Drop Rows With No Revenue
df = df[df['Amount'] > 0]
 
# Step 7: Standardize B2B Column
df['B2B'] = df['B2B'].map({True: 'B2B', False: 'B2C'}).fillna('B2C')
 
# Step 8: Fill Missing State 
df['ship-state'] = df['ship-state'].fillna('Unknown').str.title()
 
# Step 9: Clean Status Column
df['Status'] = df['Status'].str.strip()
 
# Step 10: Add Helper Columns
df['Month_Number'] = df['Date'].dt.month
df['Month_Name']   = df['Date'].dt.strftime('%b')
df['Year']         = df['Date'].dt.year
df['Quarter']      = df['Date'].dt.quarter.apply(lambda x: f'Q{x}')
 
# Step 11: Final Check & Save 
print('\nCleaned Shape:', df.shape)
print('\nData Types:')
print(df.dtypes)
df.to_csv('ecommerce_cleaned.csv', index=False)
print('\n✅ Saved: ecommerce_cleaned.csv')

```
